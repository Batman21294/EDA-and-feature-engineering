```python
country_names = final_df.Country.value_counts().index
```

```python
country_names
```

```python
country_val = final_df.Country.value_counts().values
country_val
```

```python
# pie chart - top3 countries to use zomato
plt.pie(country_val[:3],labels=country_names[:3],autopct = '%1.2f%%')
```

Observations :
Zomato's max records or transactions are from India, followed by USA and UK.

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index()
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
ratings
```

## Observations
1. when rating is between 4.5 to 4.9 --> excellent
2. when ratings are between 4.0 to 4.4 --> Very good
3. when ratings are between 3.5 to 3.9 --> Good
4. when ratings are between 2.5 to 3.4 --> Average
5. when ratings are between 1.8 to 2.4 --> Poor

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x = "Aggregate rating", y = "Rating count",data = ratings)
```

```python
sns.barplot(x = "Aggregate rating", y = "Rating count",hue = "Rating color",data = ratings,palette = ['blue','red','orange','yellow','green','green'])
```

Observation:
    1. Not rated count is very high.
    2. Max number of ratings are between 2.5 to 3.4

```python
## Count plot
sns.countplot(x = "Rating color",data = ratings,palette = ['blue','red','orange','yellow','green','green'])
# max number of orange ratings in the orange dataframe
```
